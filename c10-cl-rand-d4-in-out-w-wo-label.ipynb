```python
import pandas as pd

obj = pd.read_pickle(r'/kaggle/working/save_path/getquery_clean')
```

```python
clean_dataset = torchvision.datasets.CIFAR10(root='/content/drive/MyDrive/myCIFAR10', train = False, download = True)
```

```python
import numpy as np

correct = 0
for i in range(len(obj['0.4'])):
    
    predicted = np.argmax(obj['0.4'][i])
    
    if predicted == clean_dataset[i][1]:
        
        correct += 1
```

```python
print('clean test is = ' , (correct /  len(obj['0.4']))*100)
```
